```python
%matplotlib inline

# imports required packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pynetlogo

# think this is for the GUI idk?
sns.set_style("white")
sns.set_context("talk")

# starts a NetLogo link, point the netlogo home path to where you have it installed on your machine
netlogo = pynetlogo.NetLogoLink(
    gui=True,
    netlogo_home="/Volumes/NetLogo 6.3.0/NetLogo 6.3.0"
)

# loads a .nlogo model from provided path
netlogo.load_model("/Users/mabelwylie/Documents/NetLogo-Moss-Python/testModel.nlogo")

# runs the model setup command
netlogo.command("setup")
# runs the field allocation command
netlogo.command("allocate-fields")
```

```python
# here you can see the "soil moisture" attribute value for each of the patches, laid out similarly to netlogo
netlogo.patch_report("soil-moisture")
```

```python
netlogo.patch_report("pcolor")
```

```python
# saves the patch report for soil moisture and ground cover as dataframes
dfMoisture = netlogo.patch_report("soil-moisture")
dfCover = netlogo.patch_report("ground-cover")

for i in range(-5, 6):
    for j in range(-5, 6):
        # for each patch, calculate a new value for each attribute - this is where I envision the hydrology model coupling happening
        dfMoisture[i][j] = dfMoisture[i][j] - 10
        dfCover[i][j] = dfCover[i][j] + 10

# updates the patche attributes using the new values
netlogo.patch_set("soil-moisture", dfMoisture)
netlogo.patch_set("ground-cover", dfCover)
```
